## **Web Scraping with Beautiful Soup**

#### Using *Beautiful Soup*, you can extract a section of a webpage!

```python
from bs4 import BeautifulSoup
import requests

url = 'https://aerospace.org/article/brief-history-space-exploration'
response = requests.get(url)
html = response.text

soup = BeautifulSoup(html, 'html.parser')
div_body = soup.find('div', class_='field__body')
paragraphs = div_body.find_all('p')
text = ' '.join([p.get_text() for p in paragraphs])

print(text[:222])
```

#### In the example above the we display the first 222 characters.

## **The amazing power of *NLTK***

#### *NLTK* is a leading platform for building **Python** programs to work with human language data. It provides easy-to-use interfaces to over 50 corpora and lexical resources such as *WordNet*, along with a suite of text processing libraries for classification, tokenization, stemming, tagging, parsing, and semantic reasoning, wrappers for industrial-strength NLP libraries, and an active discussion forum.

#### By using *NLTK* we can split our text into separated words and then using *matplotlib* we can display various features such as their distribution chart.

#### Let's start with spliting the previously downloaded text into words.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

nltk.download('punkt')
nltk.download('stopwords')

words = word_tokenize(text)
sentences = sent_tokenize(text)
```

#### We can also divide the text into sentences by using the code below.

```python
sentences = sent_tokenize(text)
print(sentences[:2])
```

#### See? We can also get rid of punctuation marks!

```python
words = [word.lower() for word in words if word.isalpha()]
```

#### What is more, by using the code below, we can display the most frequent words from the text

```python
fdist = FreqDist(words)

common_words = fdist.most_common(10)
print ("Common words:")
print(common_words)
```

#### Below you can see the chart for the results.

```python
plt.title('Word Frequency Distribution')
fdist.plot(30, cumulative=False)
plt.show()
```

#### By using the code below we can get the total number of words from the text.


```python
print(len(words))
```
